## calculate correlation matrix using FastSpar program

FastSpar(github.com/scwatts/FastSpar) was used to calculate correlation matrix

## Install FastSpar:
```
conda create --name fastspar  -c bioconda -c conda-forge fastspar
```
## activate environment

```
conda activate fastspar
```

## step1: Correlation inference
```
mkdir bootstrap_counts bootstrap_correlation
```
```
fastspar --otu_table data_net.txt --correlation median_correlation.tsv --covariance median_covariance.tsv > log1
```


## step2: Calculation of exact p-values, first generate 100 bootstrap counts
```
fastspar_bootstrap --otu_table data_net.txt --number 100 --prefix bootstrap_counts/ > log2
```
## step3: infer correlations for each bootstrap count (running in parallel):
```
parallel -j 12 fastspar --otu_table {} --correlation bootstrap_correlation/cor{/} --covariance bootstrap_correlation/cov{/} -i 50 ::: bootstrap_counts/* > log3
```
## step4: From these correlations, the p-values are then calculated:
```
fastspar_pvalues --otu_table data_net.txt --correlation median_correlation.tsv --prefix bootstrap_correlation/cor --permutations 100 --outfile pvalues.tsv > log4
```

### Now back to R and select significant corralations :

---

````r
#open correlation and pvalue files and change the upper tri of the files to NA and convert them to data frame
Pvals <- read.table(file = "seqdata/BV5/pvalues.tsv", sep="\t", header=T, row.names=1,  comment.char = "")
Pvals1 <- as.matrix(Pvals)
Pvals1[upper.tri(Pvals, diag=TRUE)]<- NA
Pvals2 <- as.data.frame(as.table(Pvals1))
colnames(Pvals2)<-c("Node1","Node2","Pvalue")

head(Pvals2)
  
Cors <- read.table("seqdata/BV5/median_correlation.tsv", sep="\t", header=T, row.names=1 , comment.char = "") 
Cors1 <- as.matrix(Cors)
Cors1[upper.tri(Cors1, diag=TRUE)]<-NA
Cors2 <- as.data.frame(as.table(Cors1))
colnames(Cors2)<-c("Node1","Node2","Cor")
head(Cors2)
````

````r
#bind pvalue and correlation tables
Edge_table <- cbind(Pvals2,Cors2$Cor, deparse.level=2)
head(Edge_table)
````

````r
#remove NA lines
Edge_table_final <- Edge_table[!is.na(Edge_table$Pvalue),]
````

````r
colnames(Edge_table_final) <- c("Node1","Node2","Pvalue","Cor")

#select correlations with abs value higher than 0.2 
Edge_table_final <- subset(Edge_table_final,abs(Edge_table_final$Cor)>=0.2)
head(Edge_table_final)
````

````r
# add a new column and give it value according to correlation values
Edge_table_final$CorrB <- ifelse(Edge_table_final$Cor >=0, "Positive", "Negative")
head(Edge_table_final)
dim(Edge_table_final)
````

````r
#remove pvalues lower than 0.01
Edge_table_final_P0.01 <- Edge_table_final[which(Edge_table_final$Pvalue<=0.01),]
dim(Edge_table_final_P0.01)
#
````

````r
# number of positive and negative interactions
table(Edge_table_final_P0.01$CorrB)
````

````r
#save the file

write.table(Edge_table_final_P0.01, file="seqdata/BV5/0.01Pvalue_correlation.txt", sep="\t", row.names=FALSE , quote = F)
````

### Ploting betweeness centrality vs closeness centrality and identifying hub microbes:
---

````r
# load packages
library(ggplot2)
library(ggrepel)
````

````r
#open the analysis file
net_analysis <- read.table("seqdata/BV5/0.01Pvalue_correlation.txt default node.csv" , header = TRUE , sep = "," )
head(net_analysis)
````

````r
n <- 5 #threshold for quantile
````

````r
#select otus with high BetweennessCentrality 

top5_BC <- net_analysis[net_analysis$BetweennessCentrality > quantile(net_analysis$BetweennessCentrality,prob=1-n/100),]$name
top5_BC

````

````r
#minimum value of BetweennessCentrality for selected otus in previous step
top5_BC_min<-min(net_analysis[net_analysis$BetweennessCentrality > quantile(net_analysis$BetweennessCentrality,prob=1-n/100),]$BetweennessCentrality)
top5_BC_min
````

````r
#select otus with high ClosenessCentrality 
top5_CC<-net_analysis[net_analysis$ClosenessCentrality > quantile(net_analysis$ClosenessCentrality,prob=1-n/100),]$name

top5_CC
````

````r
#minimum value of ClosenessCentrality for selected otus in previous step
top5_CC_min<-min(net_analysis[net_analysis$ClosenessCentrality > quantile(net_analysis$ClosenessCentrality,prob=1-n/100),]$ClosenessCentrality)

top5_CC_min
````

````r
# overlap between selected otus in closeness and betweeness centralities

p5_Intersection <- top5_BC[which((top5_BC %in% top5_CC) == TRUE)]


p5_Intersection

````

````r
# consider the otus in p5_Intersection as a hub microbes and add a column in your table and name it hub. 
# if microbes are in the p5_Intersection , set the value for the hub='yes' , otherwise hub='no'

net_analysis$hub <- ifelse(net_analysis$name %in% p5_Intersection,'yes','no')
head(net_analysis)
````

````r
#plot your data
p1 = ggplot(net_analysis, aes(x = ClosenessCentrality, y= BetweennessCentrality , label=name))
p1
````

````r
p2 = p1 + geom_point(aes(colour=Degree , size = Degree) ,stat= "identity",position="identity",alpha= 0.4 ) +
scale_colour_gradient(low = "blue", high = "red")
p2
````

````r
p3 = p2 + geom_vline(xintercept = top5_CC_min, linetype="dashed")
p3
````

````r
p4 = p3 + geom_hline(yintercept = top5_BC_min, linetype= "dashed") 
p4
````

````r
p5 = p4 + geom_text_repel(data=subset(net_analysis, hub=="yes"),size = 7,
                    box.padding   = 1.5,
                    point.padding = 0.5,
                    force         = 100,
                    segment.size  = 0.4,
                    segment.color = "darkgreen")
p5
````
